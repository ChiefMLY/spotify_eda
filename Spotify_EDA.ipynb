```python
#Importing the Requred Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_df = pd.read_csv("tracks.csv")
spotify_df.head()
```

```python
#Checking for nan vlaues in the dataset
pd.isnull(spotify_df).sum()
```

```python
#checking for additional information on the dataset
spotify_df.info()
```

```python
#We will be dropping the nan rows because they are few and should be inconsequential to our analysis
spotify_clean = spotify_df.dropna(axis = 0)
spotify_clean.info()
```

```python
#10 Least Popular songs in the dataframe
sorted_df = spotify_clean.sort_values(by = 'popularity', ascending = True).head(10)
sorted_df
```

```python
# Checking the statistical summary of the dataset
spotify_clean.describe().transpose()
```

```python
#Viewing the ten top most popular songs in the data frame
most_popular1 = spotify_clean.sort_values('popularity', ascending = False).head(10)
most_popular1

# Subsetting the ten most popular songs in the data frame
most_popular = spotify_clean.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
#Setting the index to release date 
spotify_clean.set_index('release_date', inplace = True)
spotify_clean.index = pd.to_datetime(spotify_clean.index)
spotify_clean.head(10)
```

```python
#covert the duration column from miliseconds to seconds
spotify_clean['duration_s'] = spotify_clean['duration_ms'].apply(lambda x: round(x/1000))
spotify_clean.drop('duration_ms', inplace = True, axis = 1)
```

```python
spotify_clean.head()
```

```python
#View the correlation of the df
corr_df = spotify_clean.drop(['key', 'mode'], axis = 1).corr(method = 'pearson')
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = 'inferno', linewidths = 1, linecolor = 'Black')
heatmap.set_title('Corellation between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation= 90)
```

```python
#Create a sample df
SEED = 10
sample_df = spotify_clean.sample(int(0.04 * len(spotify_clean)), random_state = SEED)
sample_df.shape
```

```python
#Plotting the corr between loudness and energy using the sample df
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', scatter_kws={"color": "blue"}, line_kws={"color": "red"}).set(title = 'Loudness VS Energy')
```

```python
#Plotting the corr between popularity and acousticness using the sample df
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', scatter_kws={"color": "blue"}, line_kws={"color": "red"}).set(title = 'Popularity VS Acousticness')
```

```python
#Plotting the corr between popularity and Speechiness using the sample df
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'speechiness', scatter_kws={"color": "blue"}, line_kws={"color": "red"}).set(title = 'Popularity VS Speechiness')
```

```python
spotify_clean['dates'] =  spotify_clean.index.get_level_values('release_date')
spotify_clean.dates = pd.to_datetime(spotify_clean.dates)
spotify_clean['year'] = spotify_clean.dates.dt.year
```

```python
#View the total number of songs in each year using a distplot
sns.displot(spotify_clean['year'], discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = 'Number of songs per Year')
```

```python
#Visualizing how song duration has changed over the year
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(data = spotify_clean, x = 'year', y = 'duration_s', ax = ax, errwidth = False).set(title = 'Year Vs Duration')
plt.xticks(rotation = 90)
```

```python
fig_dims = (25,5)
fig, ax= plt.subplots(figsize = fig_dims)
fig=sns.pointplot(data = spotify_clean, x='year', y= 'duration_s', ax=ax).set(title="Year vs duration")
plt.xticks(rotation=60)
```

## Genre Analysis 
### Analysing the Spotify dataset using the Genre

```python
genre_df = pd.read_csv('SpotifyFeatures.csv')
genre_df.head(5)
```

```python
#Add a column for duration in second for easier understanding of the dataframe
genre_df['duration_s'] = genre_df.duration_ms.apply(lambda x: round(x/1000))
```

```python
#View the duration of songs in the different genres
plt.title('Duration of songs in different genres')
sns.color_palette('rocket', as_cmap = True)
sns.barplot(y = 'genre', x = 'duration_s', data = genre_df)
plt.xlabel('Duration in seconds')
plt.ylabel('Genres')
```

As observed in the barplot above there are genres that appeared more than once, so we wil have to clean it up

```python
#Viewing the top ten poular genre
sns.set_style(style = 'darkgrid')
plt.figure(figsize = (10,5))
famous = genre_df.sort_values('popularity', ascending = False).head(10)
sns.barplot(y = 'n_genre', x = 'popularity', data = famous)
plt.title('Top 5 Popular Genre')
```
